```python
# ques2-"Using the Student Grades, create a violin plot to display the distribution of scores across different grade
# categories."

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

sns.violinplot(x='Grade', y='Score', data=df, palette='muted')

plt.xlabel("Grade")
plt.ylabel("Score")
plt.title("Distribution of Scores Across Grade Categories")

plt.show()
```

```python
# ques3-" Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days"

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

pivot_table = df.pivot_table(values='Sales', index='Month', columns='Day', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".0f", cbar=True)

plt.title("Sales Variation Across Months and Days")
plt.xlabel("Day")
plt.ylabel("Month")

plt.show()
```

BOKEH ASSIGNMENT:

```python
# ques1-".Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x"

from bokeh.plotting import figure, output_file, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

output_file("sine_wave.html")

p = figure(title="Sine Wave", x_axis_label="X", y_axis_label="Y")
p.line(x, y, legend_label="Sine", line_width=2)

show(p)
```

```python
# ques2-"Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns."

from bokeh.plotting import figure, show
import numpy as np


x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(10, 50, 100)

colors = ['#%02x%02x%02x' % (r, g, b) for r, g, b in np.random.randint(0, 256, (100, 3))]

p = figure(title="Scatter Plot", x_axis_label="X", y_axis_label="Y")
p.scatter(x, y, size=sizes, color=colors, alpha=0.6)

show(p)

```

```python
# ques3-"Generate a Bokeh bar chart representing the counts of different fruits using the following dataset"

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Count")
p.vbar(x='fruits', top='counts', width=0.5, source=source)
show(p)
```

```python
# ques4-"Create a Bokeh histogram to visualize the distribution of the given data."

from bokeh.plotting import figure, show
import numpy as np

data = np.random.randn(1000)
hist, edges = np.histogram(data, bins=30)

p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black")
show(p)
```

```python
# ques5-" Create a Bokeh heatmap using the provided dataset."

from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np

data = np.random.rand(10, 10)
x = np.arange(data.shape[1])
y = np.arange(data.shape[0])

p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y")
color_mapper = LinearColorMapper(palette="Viridis256", low=data.min(), high=data.max())
p.image(image=[data], x=0, y=0, dw=10, dh=10, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
